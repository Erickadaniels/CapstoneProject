## Encoding Categorical Variable

```python
# Create a copy of the original DataFrame
df_encoded = df.copy()

# Initialize the LabelEncoder
le = LabelEncoder()

le = LabelEncoder()
for column in df.select_dtypes(include=['object']):
    df_encoded[column] = le.fit_transform(df_encoded[column])
df_encoded
```

```python
scaler = StandardScaler()
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
df_encoded[numeric_columns] = scaler.fit_transform(df[numeric_columns])
df_encoded
```

```python
negative_values = df_encoded[df_encoded < 0]
negative_values
```

## Dropping Negative Values

```python
df_cleaned = df_encoded[df_encoded >= 0].dropna()
df_cleaned
```

## Feature Engineering 

```python
df_clean = df_encoded [df_encoded  >= 0].dropna()
df_clean
```

```python
df_encoded['AvgChargesPerMonth'] = df_encoded['TotalCharges'] / df_encoded['tenure']
df_encoded['AvgChargesPerMonth']
```

```python
service_features = ['TechSupport', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'StreamingTV', 'StreamingMovies']
df_encoded['ServiceInteractions'] = df_encoded[service_features].sum(axis=1)
df_encoded['ServiceInteractions']
```

```python
# Example: Creating a new feature 'TotalTransactionValue'
df_encoded['TotalTransactionValue'] = df_cleaned['tenure'] * df_cleaned['MonthlyCharges']

# Display the new feature
df_cleaned[['tenure', 'MonthlyCharges', 'TotalTransactionValue']]
```

```python
df_encoded['tenure_with_company'] = df_encoded['tenure']*df_encoded['Contract']
df_encoded['tenure_with_company']
```

## Exploratory Data Analyis

### Key factors Contributing to Churn

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Countplot of churn distribution
sns.countplot(x='Churn', data=df_encoded)
plt.title('Distribution of Churn')
plt.show()

# Calculate the percentage of churned customers
churn_rate = df['Churn'].value_counts(normalize=True) * 100
print(f"Churn Rate:\n{churn_rate}")
```

```python
# Histograms for numerical features
numerical_features = ['tenure', 'MonthlyCharges', 'TotalCharges']

for feature in numerical_features:
    plt.figure(figsize=(10, 5))
    sns.histplot(data=df_encoded, x=feature, hue='Churn', kde=True, bins=30)
    plt.title(f'Distribution of {feature} by Churn')
    plt.show()
```

```python
# Boxplots for numerical features
for feature in numerical_features:
    plt.figure(figsize=(10, 5))
    sns.boxplot(x='Churn', y=feature, data=df_clean)
    plt.title(f'{feature} vs Churn')
    plt.show()
```
